```python
while True:
  unit = input("What unit of measurement will you use? Type 'Y' for yards of 'M' for meters: ").upper().strip()

  if unit == "Y":
    unit_name = "yards"
    break
  elif unit == "M":
    unit_name = "meters"
    break
  else:
    print("Sorry, I don't recognize that. Please type 'Y' or 'M'.")

print(f"\n--- Setting up for {unit_name} ---")

pattern_balls = int(input("How many balls of yarn does the pattern require for your size? "))
pattern_len = int(input(f"How many {unit_name} are there per ONE pattern ball? "))
sub_ball_len = int(input(f"How many {unit_name} are there in one ball of SUBSTITUTION yarn? "))

total_needed = pattern_balls * pattern_len
sub_balls_needed = -(-total_needed // sub_ball_len)

print(f"\n--- {unit_name.upper()} RESULT ---")
print(f"Total {unit_name} needed: {total_needed}")
print(f"You need {sub_balls_needed} balls of substitution yarn.")

precise_calc = total_needed / sub_ball_len
remainder = precise_calc % 1

if remainder > 0.75 or remainder == 0:
  print("TOP: You are 75% or more of the way through your last ball. Consider purchasing an extra.")
else:
  print("You should have plenty of yarn with this amount.")
```
